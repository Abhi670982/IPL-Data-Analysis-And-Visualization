```python
import pandas as pd
import numpy as np
matches = pd.read_csv('matches.csv')
matches.dtypes
matches.size
matches = matches.squeeze()
matches.dtypes

matches.values
matches.head(3)
matches.tail()
matches.describe
matches
```

What was the count of matches played in each season?

```python
matches['season'].value_counts()
## -------------------BAR Graph-------------------------
import matplotlib.pyplot as plt

seasoncount = matches['season'].value_counts().sort_index()

max_value = seasoncount.max();
colors = ['red' if value == max_value else 'Blue' for value in seasoncount ]
seasoncount.plot(kind='bar', color=colors)
plt.title('Matches Played in Each IPL Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')

plt.show()
```

How many runs were scored in each season?

```python

balls = pd.read_csv('deliveries.csv')
balls
merged = balls.merge(
    matches[['id', 'season']],
    left_on='match_id',
    right_on='id'
)
season_runs = merged.groupby('season')['total_runs'].sum()
import matplotlib.pyplot as plt

season_runs.plot(kind='line', marker='o')

plt.title('Total Runs Scored per IPL Season')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.show()
```

What were the runs scored per match in different seasons?

```python
RunsPerMatch = merged.groupby(['season', 'match_id'])['total_runs'].sum()
season_avg = RunsPerMatch.groupby('season').mean()

season_avg
season_avg.plot(kind='line', marker = 'o')
plt.title('Average of Runs per match in different seasons')
plt.xlabel('season')
plt.ylabel('total_runs')
plt.show()

```
